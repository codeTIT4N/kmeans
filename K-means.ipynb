```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans 
animals="data.csv" 
data=pd.read_csv(animals) 
data.head() 
plt.title("clustering for a1 and a2") 
plt.scatter(data.a1,data.a2) 
plt.show() 
plt.title("clustering for a3 and a4") 
plt.scatter(data.a3,data.a4) 
plt.show() 
# taking a1 as 0 a2 as 1,a3 as 2 and a4 as 3 to train my model 
x = data.iloc[:, [0,1,2,3]].values 
kmeans3 = KMeans(n_clusters=3) 
y_kmeans3 = kmeans3.fit_predict(x) 
print("\n K-mean values for n_clustering_3 is\n",y_kmeans3) 
print("\n K Means cluster for 3 clustering _centers_ is:\n",kmeans3.cluster_centers_) 
kmeans4 = KMeans(n_clusters=4) 
y_kmeans4 = kmeans4.fit_predict(x) 
print("\n K-mean values for n_clustering_4 is\n",y_kmeans4) 
print("\n K Means cluster for 4 clustering _centers_ is:\n",kmeans4.cluster_centers_) 
kmeans5 = KMeans(n_clusters=5) 
y_kmeans5 = kmeans5.fit_predict(x) 
print("\n K-mean values for n_clustering_5 is\n",y_kmeans5) 
print("\n K Means cluster for 5 clustering _centers_ is:\n",kmeans5.cluster_centers_) 
check_error =[] 
for i in range(3, 6): 
    kmeans = KMeans(n_clusters = i).fit(x) 
    kmeans.fit(x) 
    check_error.append(kmeans.inertia_) 
import matplotlib.pyplot as plt 
plt.plot(range(3, 6), check_error) 
plt.title('Elbow method') 
plt.xlabel('No of clusters') 
plt.ylabel('Error') 
plt.show() 
```
